```python
import pandas as pd

# CSV 파일 불러오기
speed_file = 'processingData.csv'  # DI_uiSpeed와 DI_vehicleSpeed가 있는 파일
torque_file = 'torquePreprocessing.csv'  # DI_torqueActual가 있는 파일

# 두 CSV 파일을 DataFrame으로 읽어오기
df_speed = pd.read_csv(speed_file)
df_torque = pd.read_csv(torque_file)

# DI_uiSpeed가 있는 DataFrame의 Timestamp를 기준으로 병합 (left join 사용)
# torqueActual 데이터는 병합 후 ffill()로 전처리
merged_df = pd.merge(df_speed, df_torque[['Timestamp', 'Physical_value']], on='Timestamp', how='left')

# DI_torqueActual이 NaN일 경우 이전 값으로 채우기
merged_df['Physical_value'] = merged_df['Physical_value'].ffill().fillna(0)

# 결과 출력
print(merged_df)

# 결과를 CSV 파일로 저장하고 싶다면 아래 코드 사용
# merged_df.to_csv('processed_data.csv', index=False)
```
